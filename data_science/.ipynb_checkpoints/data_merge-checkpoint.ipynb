### Merging two DataFrames into one

```python
import pandas as pd
```

```python
dict = {}
```

```python
pd.DataFrame({'Key1':[1,2,3,4,5]})
```

```python
df1 = pd.DataFrame({'A':[1,3,4,5,2],
                   'B':[12,5,52,15,7],
                   'C':[0,4,1,9,8],
                   'D':[101,102,130,200,122]}, index = [1,2,3,4,5])
```

```python
df1
```

```python
df2 = pd.DataFrame({'A':[7,3,2,4,1],
                   'B':[55,24,2,30,81],
                   'C':[9,41,3,12,10],
                   'D':[102,10,30,202,400]}, index = [6,7,8,9,10])
```

```python
df2
```

```python
# concat
# param needs to be list
pd.concat([df1, df2])
```

```python
# concat() merges row in default
# axis param
pd.concat([df1, df2], axis=1)
```

```python
# index needs to be set
df1.reset_index(drop=True, inplace=True)
```

```python
df1
```

```python
df2.reset_index(drop=True, inplace=True)
```

```python
df2
```

```python
pd.concat([df1,df2], axis=1)
```
