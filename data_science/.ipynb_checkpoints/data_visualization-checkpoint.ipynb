## Matplot

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,5,10)
```

```python
y = x ** 2
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y, 'r')
plt.xlabel('X_label')
plt.ylabel('Y_label')
plt.title('Title')
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.5,0.7])

axes.plot(x,y)
```

### overlapping two plots

```python
plt.plot(y, x, 'b')
plt.xlabel('X_label')
plt.ylabel('Y_label')
plt.title('Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.9, 0.9])
axes2 = fig.add_axes([0.2, 0.6, 0.3, 0.3])

axes1.plot(x, y, 'b')
plt.xlabel('X_label')
plt.ylabel('Y_label')
plt.title('Title')

axes2.plot(y, x, 'r')
plt.xlabel('X_label')
plt.ylabel('Y_label')
plt.title('Title')
```

### sub plot

```python
fig, axes = plt.subplots()
axes.plot(x,y, 'g')
```

```python
fig, axes = plt.subplots(nrows = 1, ncols = 2)
axes[0].plot(x,y,'r')
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')
axes[0].set_title('Exhibit A')

axes[1].plot(y,x, 'b')
axes[1].set_xlabel('x')
axes[1].set_ylabel('y')
axes[1].set_title('Exhibit B')
```

```python
fig.savefig('plots.png')
```

## Seaborn

```python
import seaborn as sns
```

```python
data = sns.load_dataset('tips')
```

```python
data.head()
```

### distplot

```python
sns.distplot(data['total_bill'])
```

```python
sns.distplot(data['total_bill'], kde = False)
```

```python
sns.distplot(data['total_bill'], bins = 50)
```

### jointplot

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = data)
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = data, kind = 'hex')
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = data, kind = 'reg')
```

### pairplot

```python
sns.pairplot(data)
```

```python
sns.pairplot(data, hue = 'smoker')
```

### rugplot

```python
sns.rugplot(data['total_bill'])
```

## Category plots
### barplot

```python
sns.barplot(x='smoker', y ='total_bill', data = data)
```

```python
sns.barplot(x='smoker', y ='total_bill', data = data, estimator=np.std)
#standard deviation
```

```python
sns.countplot(x='sex', data = data)
```

```python
sns.countplot(data['sex'])
```

### boxplot

```python
sns.boxplot(x='day', y ='total_bill', data = data)
```

```python
sns.boxplot(x='day', y ='total_bill', data = data, hue = 'smoker')
```

### violinplot

```python
sns.violinplot(x='day', y ='total_bill', data = data, hue = 'smoker')
```

```python
sns.violinplot(x='day', y ='total_bill', data = data, hue = 'smoker', split = True)
```

### matrixplot

```python
data.corr()
```

```python
sns.heatmap(data.corr(), cmap = 'coolwarm', annot = True)
```

### facetgrid

```python
data = sns.load_dataset('tips')
```

```python
data.head()
```

```python
g = sns.FacetGrid(data, col = 'time', row = 'smoker')
g.map(plt.hist, 'total_bill')
```

```python
g = sns.FacetGrid(data, col = 'time', row = 'smoker', hue = 'sex')
g.map(plt.scatter, 'total_bill', 'tip')
```

### regression plot
