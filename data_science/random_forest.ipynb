```python
def convertor(x):
    if x >= 0.5: return 1
    else : return 0
```

```python
# handling DataFrame
pred = pred[0].apply(lambda x: convertor(x))
```

```python
# handling npArray
pd.Series(pred).apply(lambda x: convertor(x))
```

```python
# handling array
result = []
for i in pred:
    result.append(convertor(i))
```

```python
result
```

```python
# list comporshension
result = [1 if x >= 0.5 else 0 for x in pred]
```

```python
accuracy_score(y_test, result)
```

```python
confusion_matrix(y_test, result)
```

## RandomForestRegressor vs RandomForestClassifier
[link]('https://www.quora.com/What-is-the-difference-between-scikit-learns-random-forest-classifier-and-random-forest-regressor')

- A <b>Classifier</b> is used to predict a set of specified labels. The simplest example being that of Email Spam Detection where we will always want to classify whether an email is either spam(1) or not spam(0).

- A <b>Regressor</b> is used to predict real valued outputs which vary and dont require ouptuts predicted to be in a fixed set. For example, when wanting to predict the future income of restaurants we really dont know all the possible outputs

```python
# you can shift the probability
result2 = [1 if x >= 0.3 else 0 for x in pred]
```

```python
# 0.5
print(classification_report(y_test, result))
```

```python
# 0.3
print(classification_report(y_test, result2))
```

```python
## random forest has several useful parameters
## n_estimator, default value 100, the number of created trees
rf = RandomForestRegressor(n_estimators=150, max_depth=12, random_state=100)
rf.fit(X_train, y_train)
pred = rf.predict(X_test)
result_list = [1 if x>= 0.5 else 0 for x in pred]

accuracy_score(y_test,result_list)
```

```python
rf = RandomForestRegressor(n_estimators=50, max_depth=12, random_state=100)
rf.fit(X_train, y_train)
pred = rf.predict(X_test)
result_list = [1 if x>= 0.5 else 0 for x in pred]

accuracy_score(y_test,result_list)
```
